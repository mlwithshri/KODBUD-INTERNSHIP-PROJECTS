```python
import numpy as np
import re

def convert_amount(x):
    x = str(x).lower().strip()

    # Handle unknown price
    if "call" in x:
        return np.nan

    # extract number
    nums = re.findall(r"[\d.]+", x)
    if not nums:
        return np.nan
    
    num = float(nums[0])

    if "lac" in x or "lakh" in x:
        return num * 1e5
    elif "cr" in x or "crore" in x:
        return num * 1e7

    return num
```

```python
newdf["Amount(in rupees)"] = newdf["Amount(in rupees)"].apply(convert_amount)
```
